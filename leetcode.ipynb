```python

def findMaxAverage( nums, k):
        n = len(nums)
        sum = 0


        for i in range(k):
            sum += nums[i]

        maxsum = sum

        l = 0
        r = k
        while r < n:
            sum = sum - nums[l] + nums[r]
            l += 1
            r += 1
            if sum > maxsum:
                maxsum = sum

        return maxsum / k
nums = [1,12,-5,-6,50,3]
k = 4
print(findMaxAverage( nums, k))
```

```python
def Good(s):
  count=0
  for i in s:
    if not i.isalpha():
      count+=1
    else:
      if count:
        s.remove(i)
        count-=1
      else:
        continue
  return s
s="thw$ g##bjhdwd vdg*gh$vvdd#bdudhc"
print(Good(s))
```

```python
class Solution(object):
    def recur_permute(self, nums, ds, ans, freq):
        if len(ds) == len(nums):
            ans.append(list(ds))
            return

        for i in range(len(nums)):
            if not freq[i]:
                freq[i] = True
                ds.append(nums[i])
                self.recur_permute(nums, ds, ans, freq)
                ds.pop()
                freq[i] = False

    def permute(self, nums):
        ans = []
        ds = []
        freq = [False] * len(nums)
        self.recur_permute(nums, ds, ans, freq)
        return ans
```
